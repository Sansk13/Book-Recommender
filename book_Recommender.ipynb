```python
rating=pd.read_csv('/content/drive/My Drive/recommender dataset/BX-Book-Ratings.csv',sep=';',on_bad_lines='skip',encoding='latin-1')
print(rating.head())
```

```python
rating.rename(columns={'User-ID':'userid','Book-Rating':'br'},inplace=True)
print(rating.head())
```

```python
print(rating.shape)
```

```python
rating['userid'].value_counts()
```

```python
rating['userid'].unique().shape
```

```python
x=rating['userid'].value_counts()>200
```

```python
x[x].shape
```

```python
y=x[x].index
y
```

```python
ratings=rating[rating['userid'].isin(y)]
ratings.head()
ratings.shape
```

```python
rating_with_books=ratings.merge(books,on='ISBN')
rating_with_books.head()
```

```python
num_rating=rating_with_books.groupby('title')['br'].count().reset_index()
num_rating.head()
```

```python
num_rating.shape
```

```python
num_rating.rename(columns={'br':'num_of_rating'},inplace=True)
```

```python
num_rating.head()
```

```python
final_rating=rating_with_books.merge(num_rating,on='title')
final_rating.head()
```

```python
final_rating.shape
```

```python
rating_with_books.head()
```

```python
num_rating.head()
```

```python

final_rating = final_rating.rename(columns={'num_of_rating_y': 'num_of_rating'})
final_rating.head()
```

```python
final_raitng=final_rating[final_rating['num_of_rating']>=50]
```

```python
final_raitng.head()
```

```python
final_raitng.shape
```

```python
final_raitng.drop_duplicates(['userid','title'],inplace=True)
final_raitng.shape
```

```python
book_pivot= final_raitng.pivot_table(columns='userid',index='title',values='br')
book_pivot.head()
```

```python
book_pivot.head(20)
```

```python
book_pivot.shape
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse=csr_matrix(book_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distance,suggestion=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
distance
```

```python
suggestion
```

```python
for i in range(len(suggestion)):
  print(book_pivot.index[suggestion[i]])
```

```python
book_name=book_pivot.index
```

```python
import pickle
pickle.dump(model,open('/content/drive/My Drive/recommender/model.pkl','wb'))
pickle.dump(book_name,open('/content/drive/My Drive/recommender/book_name.pkl','wb'))
pickle.dump(final_raitng,open('/content/drive/My Drive/recommender/final_rating.pkl','wb'))
pickle.dump(book_pivot,open('/content/drive/My Drive/recommender/book_pivot.pkl','wb'))
```

```python
def recommend_book(book_name):
  book_id=np.where(book_pivot.index==book_name)[0][0]
  distance,suggestion=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1),n_neighbors=10)
  for i in range(len(suggestion)):
    bookss=book_pivot.index[suggestion[i]]
  for j in bookss:
    print(j)
```

```python
recommend_book('Harry Potter and the Chamber of Secrets (Book 2)')
```
